```python
sns.catplot(
    x='departure_time',
    y='price',
    hue='departure_time',
    kind='bar',
    palette=['gold', 'red', 'blue', 'purple', 'green', 'black'],
    legend=False,
    data=data
)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
sns.catplot(
    x='arrival_time',
    y='price',
    hue='arrival_time',
    kind='bar',
    palette=['gold', 'red', 'blue', 'purple', 'green', 'black'],
    legend=False,
    data=data
)
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
sns.relplot( x = 'arrival_time', y = 'price', data = data, col = 'departure_time', kind = 'line')
plt.show()
```
